```python
print("=" * 60)
print("EXERCISE 1: RANDOM WALK PROPERTIES")
print("=" * 60)

sigma2 = 4
print(f"\nGiven: X_t = X_{{t-1}} + eps_t, eps_t ~ WN(0, {sigma2}), X_0 = 0")

print(f"\n1. Mean: E[X_t] = X_0 = 0 for all t")
print(f"2. Variance: Var(X_t) = t * sigma^2 = t * {sigma2}")

for t in [10, 50, 100]:
    print(f"   Var(X_{{{t}}}) = {t} * {sigma2} = {t*sigma2}")

print(f"\n3. Covariance: Cov(X_t, X_s) = min(t,s) * sigma^2")
print(f"   Cov(X_10, X_5) = min(10,5) * {sigma2} = {5*sigma2}")

print(f"\n4. Autocorrelation: rho(X_t, X_s) = sqrt(min(t,s)/max(t,s))")
print(f"   rho(X_10, X_5) = sqrt(5/10) = {np.sqrt(5/10):.4f}")

print(f"\n5. Key property: Var grows WITHOUT BOUND -> non-stationary")
print(f"   After differencing: Delta X_t = eps_t ~ WN(0, {sigma2}) -> stationary!")
```
