```python
print("=" * 60)
print("EXERCISE 2: DIFFERENCING PRACTICE")
print("=" * 60)

print("""
Problem 1: X_t = X_{t-1} + eps_t (Random Walk)
  Integration order: d = 1 (I(1))
  First difference: Delta X_t = eps_t ~ WN  -> Stationary

Problem 2: X_t = 2*X_{t-1} - X_{t-2} + eps_t
  Rewrite: (1 - 2L + L^2)X_t = eps_t
  Factor: (1-L)^2 X_t = eps_t
  Integration order: d = 2 (I(2))
  Need TWO differences to achieve stationarity

Problem 3: X_t = 0.5 + X_{t-1} + eps_t (RW with drift)
  Integration order: d = 1 (I(1))
  First difference: Delta X_t = 0.5 + eps_t
  After differencing: stationary with mean 0.5

Problem 4: X_t = 0.8*X_{t-1} + eps_t (AR(1))
  |phi| = 0.8 < 1 -> Already stationary!
  Integration order: d = 0 (I(0))
  No differencing needed. This is ARIMA(1,0,0) = AR(1)
""")

print("Rule of thumb:")
print("  ACF decays quickly -> stationary (d=0)")
print("  ACF decays very slowly -> difference once (d=1)")
print("  After d=1, ACF still slow -> try d=2")
print("  WARNING: Over-differencing introduces negative autocorrelation!")
```
