```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import acf
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
COLORS = {'blue': '#1A3A6E', 'red': '#DC3545', 'green': '#2E7D32', 'orange': '#E67E22', 'gray': '#666666', 'purple': '#8E44AD'}
BLUE, RED, GREEN, ORANGE, GRAY, PURPLE = COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['orange'], COLORS['gray'], COLORS['purple']
plt.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none', 'savefig.facecolor': 'none',
    'savefig.transparent': True, 'axes.spines.top': False, 'axes.spines.right': False,
    'axes.grid': False, 'font.size': 10, 'axes.titlesize': 12, 'axes.labelsize': 10,
    'xtick.labelsize': 9, 'ytick.labelsize': 9, 'legend.fontsize': 9, 'figure.dpi': 150,
    'lines.linewidth': 1.2, 'axes.linewidth': 0.6, 'legend.facecolor': 'none',
    'legend.framealpha': 0, 'legend.edgecolor': 'none',
})
def save_chart(fig, name):
    fig.savefig(f'{name}.pdf', bbox_inches='tight', transparent=True, dpi=150)
    fig.savefig(f'{name}.png', bbox_inches='tight', transparent=True, dpi=150)
    print(f'Saved: {name}')
```

```python
np.random.seed(42); n = 500
d = 0.35; e = np.random.normal(0, 0.005, n); rate = np.zeros(n); rate[0] = 4.95
for t in range(1, n):
    weights = [d/k for k in range(1, min(t+1, 30))]
    rate[t] = rate[t-1] + sum(w*(rate[max(0,t-i-1)] - rate[max(0,t-i-2)]) for i, w in enumerate(weights)) + e[t]
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(rate, color=BLUE, lw=0.8); axes[0].set_title('EUR/RON Exchange Rate (Simulated)', fontweight='bold')
acf_vals = acf(np.diff(np.abs(np.diff(rate))), nlags=50)
axes[1].bar(range(len(acf_vals)), acf_vals, color=BLUE, width=0.5)
axes[1].set_title('ACF: Slow Hyperbolic Decay', fontweight='bold')
plt.tight_layout(); save_chart(fig, 'ch8_eurron_long_memory'); plt.show()
```
