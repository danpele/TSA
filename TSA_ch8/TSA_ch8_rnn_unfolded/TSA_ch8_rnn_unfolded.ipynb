```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import acf
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
COLORS = {'blue': '#1A3A6E', 'red': '#DC3545', 'green': '#2E7D32', 'orange': '#E67E22', 'gray': '#666666', 'purple': '#8E44AD'}
BLUE, RED, GREEN, ORANGE, GRAY, PURPLE = COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['orange'], COLORS['gray'], COLORS['purple']
plt.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none', 'savefig.facecolor': 'none',
    'savefig.transparent': True, 'axes.spines.top': False, 'axes.spines.right': False,
    'axes.grid': False, 'font.size': 10, 'axes.titlesize': 12, 'axes.labelsize': 10,
    'xtick.labelsize': 9, 'ytick.labelsize': 9, 'legend.fontsize': 9, 'figure.dpi': 150,
    'lines.linewidth': 1.2, 'axes.linewidth': 0.6, 'legend.facecolor': 'none',
    'legend.framealpha': 0, 'legend.edgecolor': 'none',
})
def save_chart(fig, name):
    fig.savefig(f'{name}.pdf', bbox_inches='tight', transparent=True, dpi=150)
    fig.savefig(f'{name}.png', bbox_inches='tight', transparent=True, dpi=150)
    print(f'Saved: {name}')
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
steps = 5
for t in range(steps):
    x = t * 2.0
    ax.add_patch(plt.Rectangle((x-0.3, 0.3), 0.6, 0.6, fill=True, facecolor=BLUE, alpha=0.3, edgecolor=BLUE, lw=2))
    ax.text(x, 0.6, f'$h_{t}$', ha='center', va='center', fontsize=12, fontweight='bold')
    ax.text(x, -0.1, f'$x_{t}$', ha='center', va='center', fontsize=10)
    ax.annotate('', xy=(x, 0.3), xytext=(x, 0.05), arrowprops=dict(arrowstyle='->', color=GREEN, lw=1.5))
    ax.text(x, 1.2, f'$y_{t}$', ha='center', va='center', fontsize=10)
    ax.annotate('', xy=(x, 1.05), xytext=(x, 0.9), arrowprops=dict(arrowstyle='->', color=RED, lw=1.5))
    if t < steps-1:
        ax.annotate('', xy=((t+1)*2-0.3, 0.6), xytext=(x+0.3, 0.6), arrowprops=dict(arrowstyle='->', color=ORANGE, lw=2))
ax.set_xlim(-1, steps*2); ax.set_ylim(-0.5, 1.5); ax.axis('off')
ax.set_title('RNN Unfolded in Time', fontweight='bold', fontsize=14)
plt.tight_layout(); save_chart(fig, 'ch8_rnn_unfolded'); plt.show()
```
