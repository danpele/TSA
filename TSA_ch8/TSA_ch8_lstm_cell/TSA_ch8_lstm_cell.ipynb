```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import acf
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
COLORS = {'blue': '#1A3A6E', 'red': '#DC3545', 'green': '#2E7D32', 'orange': '#E67E22', 'gray': '#666666', 'purple': '#8E44AD'}
BLUE, RED, GREEN, ORANGE, GRAY, PURPLE = COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['orange'], COLORS['gray'], COLORS['purple']
plt.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none', 'savefig.facecolor': 'none',
    'savefig.transparent': True, 'axes.spines.top': False, 'axes.spines.right': False,
    'axes.grid': False, 'font.size': 10, 'axes.titlesize': 12, 'axes.labelsize': 10,
    'xtick.labelsize': 9, 'ytick.labelsize': 9, 'legend.fontsize': 9, 'figure.dpi': 150,
    'lines.linewidth': 1.2, 'axes.linewidth': 0.6, 'legend.facecolor': 'none',
    'legend.framealpha': 0, 'legend.edgecolor': 'none',
})
def save_chart(fig, name):
    fig.savefig(f'{name}.pdf', bbox_inches='tight', transparent=True, dpi=150)
    fig.savefig(f'{name}.png', bbox_inches='tight', transparent=True, dpi=150)
    print(f'Saved: {name}')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
rect = plt.Rectangle((0.1, 0.1), 0.8, 0.8, fill=False, color=BLUE, lw=2, ls='--')
ax.add_patch(rect)
# Gates
for x, label, c in [(0.25, 'Forget\nGate', RED), (0.5, 'Input\nGate', GREEN), (0.75, 'Output\nGate', ORANGE)]:
    ax.add_patch(plt.Circle((x, 0.7), 0.08, color=c, alpha=0.8))
    ax.text(x, 0.55, label, ha='center', va='center', fontsize=8, fontweight='bold')
# Memory cell
ax.add_patch(plt.Rectangle((0.3, 0.25), 0.4, 0.15, color=PURPLE, alpha=0.3))
ax.text(0.5, 0.325, 'Cell State $c_t$', ha='center', va='center', fontsize=10, fontweight='bold')
# Arrows
ax.annotate('$h_{t-1}$', xy=(0.1, 0.5), fontsize=10, ha='center')
ax.annotate('$h_t$', xy=(0.9, 0.5), fontsize=10, ha='center')
ax.annotate('$x_t$', xy=(0.5, 0.05), fontsize=10, ha='center')
ax.set_xlim(0, 1); ax.set_ylim(0, 1); ax.set_aspect('equal'); ax.axis('off')
ax.set_title('LSTM Cell Architecture', fontweight='bold', fontsize=14)
plt.tight_layout(); save_chart(fig, 'ch8_lstm_cell'); plt.show()
```
