```python
np.random.seed(42); n = 2000
returns = np.random.normal(0, 1, n); vol = np.ones(n)
for t in range(1, n): vol[t] = 0.02 + 0.08*returns[t-1]**2 + 0.9*vol[t-1]
returns = returns * np.sqrt(vol)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(returns[-500:], color=BLUE, lw=0.3); axes[0].set_title('S&P 500 Returns (Simulated)', fontweight='bold')
acf_abs = acf(np.abs(returns), nlags=100)
axes[1].bar(range(len(acf_abs)), acf_abs, color=RED, width=0.5)
axes[1].set_title('ACF: |Returns| - Long Memory', fontweight='bold')
plt.tight_layout(); save_chart(fig, 'ch8_arfima_real_data'); plt.show()
```
