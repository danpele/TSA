```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_morlet_wavelet
# Morlet wavelet at multiple scales: real part, envelope, Fourier transform
def morlet(t, omega0=6.0, scale=1.0):
    ts = t / scale
    return (np.pi**(-0.25) * np.exp(1j*omega0*ts) * np.exp(-ts**2/2)) / np.sqrt(scale)

t = np.linspace(-8, 8, 1000)
omega0 = 6.0
scales = [0.5, 1.0, 2.0, 4.0]
cols   = [COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['purple']]

fig, axes = plt.subplots(2, 2, figsize=(11, 7))

for scale, col in zip(scales, cols):
    w = morlet(t, omega0, scale)
    axes[0, 0].plot(t, np.real(w), color=col, lw=1.5, alpha=0.85, label=f'scale={scale}')
axes[0, 0].set_title('Morlet Wavelet — Real Part')
axes[0, 0].set_xlabel('Time t'); axes[0, 0].set_ylabel('Amplitude')
axes[0, 0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4, frameon=False)

w_single = morlet(t, omega0, 1.0)
freqs = np.fft.rfftfreq(len(t), d=t[1]-t[0])
W = np.abs(np.fft.rfft(np.real(w_single)))
axes[0, 1].plot(freqs, W, color=COLORS['blue'], lw=1.8)
axes[0, 1].set_title('Fourier Transform of Morlet (scale=1)')
axes[0, 1].set_xlabel('Frequency'); axes[0, 1].set_ylabel('|W(f)|'); axes[0, 1].set_xlim(0, 3)

for scale, col in zip(scales, cols):
    w = morlet(t, omega0, scale)
    axes[1, 0].plot(t, np.abs(w), color=col, lw=1.5, alpha=0.85, label=f'scale={scale}')
axes[1, 0].set_title('Morlet Wavelet — Envelope')
axes[1, 0].set_xlabel('Time t'); axes[1, 0].set_ylabel('Amplitude')
axes[1, 0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4, frameon=False)

axes[1, 1].plot(t, np.imag(morlet(t, omega0, 1.0)), color=COLORS['orange'], lw=1.5, label='Imaginary part')
axes[1, 1].plot(t, np.abs(morlet(t, omega0, 1.0)), color=COLORS['red'], lw=1.5, linestyle='--', label='Envelope')
axes[1, 1].set_title('Morlet (scale=1) — Imaginary & Envelope')
axes[1, 1].set_xlabel('Time t'); axes[1, 1].set_ylabel('Amplitude')
axes[1, 1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2, frameon=False)

for ax in axes.flat:
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('ch12_morlet_wavelet.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_morlet_wavelet.png', bbox_inches='tight', dpi=150)
plt.show()
```
