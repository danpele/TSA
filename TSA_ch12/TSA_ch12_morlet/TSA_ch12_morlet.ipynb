**Description:** Morlet wavelet and continuous wavelet transform (CWT)

```python
import numpy as np
import matplotlib.pyplot as plt

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
def morlet(t, omega0=6.0, scale=1.0):
    """Morlet wavelet at given scale."""
    ts = t / scale
    return (np.pi**(-0.25) * np.exp(1j*omega0*ts) * np.exp(-ts**2/2)) / np.sqrt(scale)

t = np.linspace(-8, 8, 1000)
omega0 = 6.0

fig, axes = plt.subplots(2, 2, figsize=(11, 7))
scales = [0.5, 1.0, 2.0, 4.0]
cols   = [COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['purple']]

for ax_row, part, title in zip([axes[0], axes[1]], ['real', 'envelope'],
                                ['Real Part', 'Envelope']):
    for scale, col in zip(scales, cols):
        w = morlet(t, omega0, scale)
        if part == 'real':
            ax_row[0].plot(t, np.real(w), color=col, lw=1.5, alpha=0.85, label=f'scale={scale}')
        else:
            ax_row[0].plot(t, np.abs(w), color=col, lw=1.5, alpha=0.85, label=f'scale={scale}')
    ax_row[0].set_title(f'Morlet Wavelet — {title}')
    ax_row[0].set_xlabel('Time t')
    ax_row[0].set_ylabel('Amplitude')
    ax_row[0].spines[['top','right']].set_visible(False)
    ax_row[0].set_facecolor('none')
    ax_row[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4, frameon=False)

# Fourier transform of Morlet
w_single = morlet(t, omega0, 1.0)
freqs = np.fft.rfftfreq(len(t), d=t[1]-t[0])
W = np.abs(np.fft.rfft(np.real(w_single)))
axes[0, 1].plot(freqs, W, color=COLORS['blue'], lw=1.8)
axes[0, 1].set_title('Fourier Transform of Morlet (scale=1)')
axes[0, 1].set_xlabel('Frequency')
axes[0, 1].set_ylabel('|W(f)|')
axes[0, 1].set_xlim(0, 3)
axes[0, 1].spines[['top','right']].set_visible(False)
axes[0, 1].set_facecolor('none')

# Imaginary part
axes[1, 1].plot(t, np.imag(morlet(t, omega0, 1.0)), color=COLORS['orange'], lw=1.5, label='Imaginary part')
axes[1, 1].plot(t, np.abs(morlet(t, omega0, 1.0)), color=COLORS['red'], lw=1.5, linestyle='--', label='Envelope')
axes[1, 1].set_title('Morlet Wavelet — Imaginary Part & Envelope (scale=1)')
axes[1, 1].set_xlabel('Time t')
axes[1, 1].set_ylabel('Amplitude')
axes[1, 1].spines[['top','right']].set_visible(False)
axes[1, 1].set_facecolor('none')
axes[1, 1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2, frameon=False)

fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('morlet_wavelet.pdf', bbox_inches='tight')
plt.show()
```
