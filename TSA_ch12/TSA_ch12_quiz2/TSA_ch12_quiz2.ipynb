```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram, welch
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_quiz2_inconsistency
# Three replicates of periodogram vs Welch estimator
rng = np.random.default_rng(99)
phi = 0.6
N = 256
true_psd = lambda f: 1.0 / (1 - 2*phi*np.cos(2*np.pi*f) + phi**2)
f_t = np.linspace(0.01, 0.49, 300)

fig, axes = plt.subplots(1, 4, figsize=(15, 4))
for i in range(3):
    eps = rng.normal(0, 1, N + 50)
    x = np.zeros(N + 50)
    for t in range(1, N+50): x[t] = phi*x[t-1] + eps[t]
    x = x[50:]
    f, p = periodogram(x)
    axes[i].semilogy(f[1:], p[1:], color=COLORS['blue'], lw=0.9, alpha=0.8, label=f'Rep {i+1}')
    axes[i].semilogy(f_t, true_psd(f_t), 'k--', lw=1.5, label='True')
    axes[i].set_title(f'Periodogram — Rep {i+1}')
    axes[i].set_xlabel('Frequency'); axes[i].set_ylabel('PSD')
    axes[i].spines[['top','right']].set_visible(False)
    axes[i].set_facecolor('none')
    axes[i].legend(loc='upper center', bbox_to_anchor=(0.5,-0.22), ncol=1, frameon=False)

eps = rng.normal(0, 1, N + 50)
x = np.zeros(N + 50)
for t in range(1, N+50): x[t] = phi*x[t-1] + eps[t]
x = x[50:]
f_w, p_w = welch(x, nperseg=64)
axes[3].semilogy(f_w[1:], p_w[1:], color=COLORS['red'], lw=1.8, label='Welch')
axes[3].semilogy(f_t, true_psd(f_t), 'k--', lw=1.5, label='True')
axes[3].set_title('Welch Estimator')
axes[3].set_xlabel('Frequency')
axes[3].spines[['top','right']].set_visible(False)
axes[3].set_facecolor('none')
axes[3].legend(loc='upper center', bbox_to_anchor=(0.5,-0.22), ncol=1, frameon=False)

fig.patch.set_alpha(0)
plt.suptitle('Periodogram Inconsistency (N=256): Welch reduces variance by averaging sub-periodograms', y=1.04)
plt.tight_layout()
plt.savefig('ch12_quiz2_inconsistency.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_quiz2_inconsistency.png', bbox_inches='tight', dpi=150)
plt.show()
```
