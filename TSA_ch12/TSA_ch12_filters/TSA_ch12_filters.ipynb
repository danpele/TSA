```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_filter_transfer
# HP, BK and CF gain functions in the frequency domain
omega = np.linspace(0, np.pi, 2000)

def hp_transfer(omega, lam):
    return (4*lam*(1 - np.cos(omega))**2 / (1 + 4*lam*(1 - np.cos(omega))**2))**2

def ideal_bandpass(omega, pl, pu):
    return ((omega >= 2*np.pi/pu) & (omega <= 2*np.pi/pl)).astype(float)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(omega/np.pi, hp_transfer(omega, 1600)**0.5, color=COLORS['blue'],  lw=2, label='HP λ=1600')
bk = ideal_bandpass(omega, 6, 32)
ax.plot(omega/np.pi, bk, color=COLORS['green'], lw=2, linestyle='--', label='BK ideal (6–32q)')
cf = ideal_bandpass(omega, 6, 32)
ax.plot(omega/np.pi, cf * 0.95, color=COLORS['red'], lw=2, linestyle=':', label='CF ideal (6–32q)')
ax.axvspan(2/32, 2/6, alpha=0.12, color=COLORS['orange'], label='Business cycle band')
ax.set_xlabel('Frequency (×π rad/sample)')
ax.set_ylabel('Gain')
ax.set_title('Transfer Functions: HP, BK, CF Filters')
ax.set_xlim(0, 0.6)
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=4, frameon=False)
plt.tight_layout()
plt.savefig('ch12_filter_transfer.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_filter_transfer.png', bbox_inches='tight', dpi=150)
plt.show()
```
