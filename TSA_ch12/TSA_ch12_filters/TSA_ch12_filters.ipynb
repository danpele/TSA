**Description:** Linear filter transfer functions — HP, BK, and CF filters

```python
import numpy as np
import matplotlib.pyplot as plt

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
omega = np.linspace(0, np.pi, 2000)

def hp_transfer(omega, lam):
    """HP filter gain squared |A_HP(omega)|^2."""
    return (4*lam*(1 - np.cos(omega))**2 / (1 + 4*lam*(1 - np.cos(omega))**2))**2

def bk_transfer(omega, pl=6, pu=32, K=12):
    """BK band-pass approximation (ideal band-pass, no end-point correction)."""
    return ((np.abs(omega) >= 2*np.pi/pu) & (np.abs(omega) <= 2*np.pi/pl)).astype(float)

def ideal_bandpass(omega, pl, pu):
    return ((omega >= 2*np.pi/pu) & (omega <= 2*np.pi/pl)).astype(float)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(omega/np.pi, hp_transfer(omega, 1600)**0.5, color=COLORS['blue'],  lw=2, label='HP λ=1600')
ax.fill_between([6/32, 1], 0, 1.05, alpha=0.08, color=COLORS['amber'])
ax.axvspan(2*np.pi/(32*np.pi), 2*np.pi/(6*np.pi), alpha=0.08, color=COLORS['amber'])

# BK ideal band (6–32 quarters)
bk = ideal_bandpass(omega, 6, 32)
ax.plot(omega/np.pi, bk, color=COLORS['green'], lw=2, linestyle='--', label='BK ideal (6–32q)')

# CF ideal band
cf = ideal_bandpass(omega, 6, 32)
ax.plot(omega/np.pi, cf * 0.95, color=COLORS['red'], lw=2, linestyle=':', label='CF ideal (6–32q)')

# Business cycle band shading
ax.axvspan(2/(32), 2/(6), alpha=0.12, color=COLORS['amber'], label='Business cycle band')
ax.set_xlabel('Frequency (×π rad/sample = cycles/2 periods)')
ax.set_ylabel('Gain')
ax.set_title('Transfer Functions: HP, BK, CF Filters')
ax.set_xlim(0, 0.6)
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=4, frameon=False)
plt.tight_layout()
plt.savefig('filter_transfer.pdf', bbox_inches='tight')
plt.show()
```
