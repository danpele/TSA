```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_quiz3_seasonal_spectrum
# Periodogram of a seasonal series: identify spike at f=1/12
rng = np.random.default_rng(5)
N = 240
t = np.arange(N)
season_period = 12
x = np.sin(2*np.pi*t/season_period) + 0.5*np.sin(2*np.pi*t/6) + 0.3*rng.normal(size=N)

f, p = periodogram(x, fs=1.0)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(t, x, color=COLORS['blue'], lw=1)
axes[0].set_title('Simulated Seasonal Series (T=12 months)')
axes[0].set_xlabel('Time (months)'); axes[0].set_ylabel('Value')

axes[1].plot(f[1:N//2], p[1:N//2], color=COLORS['blue'], lw=1)
axes[1].axvline(1/12, color=COLORS['red'],   lw=1.8, linestyle='--', label='f=1/12 (T=12)')
axes[1].axvline(1/6,  color=COLORS['green'], lw=1.8, linestyle='--', label='f=1/6  (T=6)')
axes[1].set_title('Periodogram — Seasonal Spikes')
axes[1].set_xlabel('Frequency (cycles/month)'); axes[1].set_ylabel('Power')
axes[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=2, frameon=False)

for ax in axes:
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('ch12_quiz3_seasonal_spectrum.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_quiz3_seasonal_spectrum.png', bbox_inches='tight', dpi=150)
plt.show()
print('Spike at f=1/12 → period = 12 months; ω = 2π/12 = π/6 rad/sample')
```
