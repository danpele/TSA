**Description:** Sunspot cycle analysis — periodogram reveals 11-year cycle

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.signal import periodogram

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
# Load sunspot data
data = sm.datasets.sunspots.load_pandas().data
years = data['YEAR'].values
activity = data['SUNACTIVITY'].values

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(years, activity, color=COLORS['blue'], lw=1)
ax.set_xlabel('Year')
ax.set_ylabel('Sunspot Activity')
ax.set_title('Annual Sunspot Numbers')
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('sunspot_timeseries.pdf', bbox_inches='tight')
plt.show()
```

```python
# Periodogram of sunspot series
fs = 1.0  # 1 observation per year
freqs, psd = periodogram(activity, fs=fs)

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(freqs[1:], psd[1:], color=COLORS['blue'], lw=1, label='Periodogram')
cycle_freq = 1/11
ax.axvline(cycle_freq, color=COLORS['red'], lw=1.5, linestyle='--', label='11-year cycle (1/11 yr⁻¹)')
ax.set_xlabel('Frequency (cycles/year)')
ax.set_ylabel('Power Spectral Density')
ax.set_title('Periodogram of Sunspot Activity')
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=2, frameon=False)
plt.tight_layout()
plt.savefig('sunspot_periodogram.pdf', bbox_inches='tight')
plt.show()
print(f'Dominant period: {1/freqs[1:][psd[1:].argmax()]:.1f} years')
```
