```python
import os
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.signal import periodogram
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_sunspot_timeseries
# Annual sunspot numbers: time series and periodogram showing 11-year cycle
data = sm.datasets.sunspots.load_pandas().data
years = data['YEAR'].values
activity = data['SUNACTIVITY'].values

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].plot(years, activity, color=COLORS['blue'], lw=1)
axes[0].set_xlabel('Year'); axes[0].set_ylabel('Sunspot Activity')
axes[0].set_title('Annual Sunspot Numbers')
axes[0].spines[['top','right']].set_visible(False)
axes[0].set_facecolor('none')

fs = 1.0
freqs, psd = periodogram(activity, fs=fs)
axes[1].plot(freqs[1:], psd[1:], color=COLORS['blue'], lw=1, label='Periodogram')
cycle_freq = 1/11
axes[1].axvline(cycle_freq, color=COLORS['red'], lw=1.5, linestyle='--', label='11-year cycle (1/11 yr⁻¹)')
axes[1].set_xlabel('Frequency (cycles/year)'); axes[1].set_ylabel('Power Spectral Density')
axes[1].set_title('Periodogram of Sunspot Activity')
axes[1].spines[['top','right']].set_visible(False)
axes[1].set_facecolor('none')
axes[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=2, frameon=False)

fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('ch12_sunspot_timeseries.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_sunspot_timeseries.png', bbox_inches='tight', dpi=150)
plt.show()
dominant_period = 1 / freqs[1:][psd[1:].argmax()]
print(f'Dominant period: {dominant_period:.1f} years')
```
