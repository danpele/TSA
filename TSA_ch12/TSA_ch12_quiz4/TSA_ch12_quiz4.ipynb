**Description:** Quiz 4 — wavelet advantage

**Question:** Why do wavelets outperform Fourier for non-stationary series?

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
N = 512
t = np.linspace(0, 10, N)
# Two-regime signal: low freq in [0,5), high freq in [5,10]
f_low, f_high = 0.5, 2.0
x = np.where(t < 5,
             np.sin(2*np.pi*f_low*t),
             np.sin(2*np.pi*f_high*t))

# Fourier spectrum (cannot localise)
X = np.fft.rfft(x)
freqs_fft = np.fft.rfftfreq(N, d=t[1]-t[0])

# Wavelet scalogram
wavelet = 'cmor1.5-1.0'
scales = np.geomspace(2, 128, 60)
coeffs, freqs_cwt = pywt.cwt(x, scales, wavelet, sampling_period=t[1]-t[0])
power = np.abs(coeffs)**2

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].plot(t, x, color=COLORS['blue'], lw=1)
axes[0].axvline(5, color=COLORS['red'], lw=1.5, linestyle='--', label='Regime change')
axes[0].set_title('Non-stationary Signal')
axes[0].set_xlabel('Time'); axes[0].set_ylabel('x(t)')
axes[0].legend(frameon=False)

axes[1].plot(freqs_fft, np.abs(X)/N, color=COLORS['blue'], lw=1.5)
axes[1].axvline(f_low,  color=COLORS['green'], lw=1.5, linestyle='--', label=f'f_low={f_low}')
axes[1].axvline(f_high, color=COLORS['red'],   lw=1.5, linestyle='--', label=f'f_high={f_high}')
axes[1].set_title('Fourier Spectrum (no time info)')
axes[1].set_xlabel('Frequency (Hz)'); axes[1].set_ylabel('|X(f)|/N')
axes[1].set_xlim(0, 4)
axes[1].legend(frameon=False, fontsize=8)

im = axes[2].contourf(t, freqs_cwt, power, levels=25, cmap='YlOrRd')
axes[2].axvline(5, color='white', lw=1.5, linestyle='--')
axes[2].set_title('Wavelet Scalogram (time–frequency)')
axes[2].set_xlabel('Time'); axes[2].set_ylabel('Frequency (Hz)')
axes[2].set_yscale('log')
plt.colorbar(im, ax=axes[2], label='Power')

for ax in axes:
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('quiz4_wavelet_advantage.pdf', bbox_inches='tight')
plt.show()
```
