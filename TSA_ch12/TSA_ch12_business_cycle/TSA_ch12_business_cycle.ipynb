**Description:** Business cycle analysis using spectral methods

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.filters.hp_filter import hpfilter
from statsmodels.tsa.filters.bk_filter import bkfilter

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
data = sm.datasets.macrodata.load_pandas().data
gdp = np.log(data['realgdp'].values)
year_q = data['year'].values + (data['quarter'].values - 1) / 4

cycle_hp, _ = hpfilter(gdp, lamb=1600)
cycle_bk = bkfilter(gdp, low=6, high=32, K=12)
T = len(gdp)
n_bk = len(cycle_bk)
start_bk = (T - n_bk) // 2

# Approximate NBER recession dates (year ranges)
recessions = [(1960.5, 1961.1), (1969.9, 1970.9), (1973.9, 1975.2),
              (1980.0, 1980.5), (1981.5, 1982.9), (1990.5, 1991.2),
              (2001.2, 2001.9), (2007.9, 2009.5)]

fig, ax = plt.subplots(figsize=(12, 4))
for (start, end) in recessions:
    ax.axvspan(start, end, color=COLORS['gray'], alpha=0.25)

ax.plot(year_q, cycle_hp, color=COLORS['blue'], lw=1.5, label='HP cycle (λ=1600)')
ax.plot(year_q[start_bk:start_bk+n_bk], cycle_bk,
        color=COLORS['red'], lw=1.5, label='BK cycle (6–32q)', linestyle='--')
ax.axhline(0, color='black', lw=0.7)
ax.set_title('US Business Cycles: HP vs BK (shaded = NBER recessions)')
ax.set_xlabel('Year')
ax.set_ylabel('Cycle component (log GDP)')
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=False)
plt.tight_layout()
plt.savefig('business_cycle.pdf', bbox_inches='tight')
plt.show()
```
