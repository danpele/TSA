**Description:** Spectral leakage and data tapering with window functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram
from scipy.signal.windows import hann

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
N = 128
t = np.arange(N)

# Grid frequency (no leakage)
f_grid = 10 / N
# Non-grid frequency (leakage)
f_nongrid = 10.5 / N

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

for ax, f, title in zip(axes,
                         [f_grid, f_nongrid],
                         ['Grid frequency (f=10/N)', 'Non-grid frequency (f=10.5/N)']):
    x = np.sin(2*np.pi*f*t)
    # No window
    fr, p_rect = periodogram(x)
    # Hann window
    w = hann(N)
    fr_h, p_hann = periodogram(x * w / np.sqrt(np.mean(w**2)))
    ax.semilogy(fr, p_rect + 1e-12, color=COLORS['blue'], lw=1.3, label='Rectangular')
    ax.semilogy(fr_h, p_hann + 1e-12, color=COLORS['red'], lw=1.3,
                linestyle='--', label='Hann window')
    ax.set_title(title)
    ax.set_xlabel('Frequency')
    ax.set_ylabel('PSD (log)')
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=2, frameon=False)

fig.patch.set_alpha(0)
plt.suptitle('Spectral Leakage: Effect of Windowing')
plt.tight_layout()
plt.savefig('spectral_leakage.pdf', bbox_inches='tight')
plt.show()
```
