```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_quiz5_hp_filter
# HP filter gain for different lambda values: cutoff frequency shifts
omega = np.linspace(1e-3, np.pi, 2000)

def hp_gain_sq(omega, lam):
    return (4*lam*(1 - np.cos(omega))**2)**2 / (1 + 4*lam*(1 - np.cos(omega))**2)**2

lambdas = [100, 1600, 14400]
cols    = [COLORS['blue'], COLORS['red'], COLORS['green']]
fig, ax = plt.subplots(figsize=(10, 5))

for lam, col in zip(lambdas, cols):
    gain = hp_gain_sq(omega, lam)
    ax.plot(omega/np.pi, gain, color=col, lw=2, label=f'λ = {lam:,}')
    idx = np.argmin(np.abs(gain - 0.5))
    ax.axvline(omega[idx]/np.pi, color=col, lw=1, linestyle=':')
    period_cutoff = 2*np.pi / omega[idx]
    print(f'λ={lam:5d}: cutoff ω/π ≈ {omega[idx]/np.pi:.3f}, period ≈ {period_cutoff:.1f} quarters')

ax.axhline(0.5, color='black', lw=0.8, linestyle='--', alpha=0.5, label='-3 dB cutoff')
ax.set_xlabel('Frequency (×π rad/sample)')
ax.set_ylabel('|A_HP(ω)|²  (gain squared)')
ax.set_title('HP Filter Transfer Function for Different λ')
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=4, frameon=False)
plt.tight_layout()
plt.savefig('ch12_quiz5_hp_filter.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_quiz5_hp_filter.png', bbox_inches='tight', dpi=150)
plt.show()
```
