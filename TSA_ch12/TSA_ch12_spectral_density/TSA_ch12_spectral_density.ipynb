**Description:** Spectral density functions of ARMA processes

```python
import numpy as np
import matplotlib.pyplot as plt

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
omega = np.linspace(0, np.pi, 500)
sigma2 = 1.0

def ar1_spectrum(phi, omega):
    return sigma2 / (1 - 2*phi*np.cos(omega) + phi**2)

def ma1_spectrum(theta, omega):
    return sigma2 * (1 + 2*theta*np.cos(omega) + theta**2)

fig, axes = plt.subplots(2, 2, figsize=(11, 7))

# AR(1) phi = 0.3, 0.6, 0.9
for phi, col in [(0.3, COLORS['blue']), (0.6, COLORS['orange']), (0.9, COLORS['red'])]:
    axes[0,0].plot(omega/np.pi, ar1_spectrum(phi, omega), color=col, lw=1.8, label=f'φ={phi}')
axes[0,0].set_title('AR(1) Spectral Density')
axes[0,0].set_xlabel('Frequency (×π rad/sample)')
axes[0,0].set_ylabel('f(ω)')

# MA(1) theta = +0.5, -0.5
for theta, col, ls in [(0.5, COLORS['green'], '-'), (-0.5, COLORS['purple'], '--')]:
    axes[0,1].plot(omega/np.pi, ma1_spectrum(theta, omega), color=col, lw=1.8,
                   linestyle=ls, label=f'θ={theta}')
axes[0,1].set_title('MA(1) Spectral Density')
axes[0,1].set_xlabel('Frequency (×π rad/sample)')
axes[0,1].set_ylabel('f(ω)')

# AR(1) log scale
for phi, col in [(0.3, COLORS['blue']), (0.6, COLORS['orange']), (0.9, COLORS['red'])]:
    axes[1,0].semilogy(omega/np.pi, ar1_spectrum(phi, omega), color=col, lw=1.8, label=f'φ={phi}')
axes[1,0].set_title('AR(1) — Log Scale')
axes[1,0].set_xlabel('Frequency (×π rad/sample)')
axes[1,0].set_ylabel('log f(ω)')

# MA(1) log scale
for theta, col, ls in [(0.5, COLORS['green'], '-'), (-0.5, COLORS['purple'], '--')]:
    axes[1,1].semilogy(omega/np.pi, ma1_spectrum(theta, omega), color=col, lw=1.8,
                       linestyle=ls, label=f'θ={theta}')
axes[1,1].set_title('MA(1) — Log Scale')
axes[1,1].set_xlabel('Frequency (×π rad/sample)')
axes[1,1].set_ylabel('log f(ω)')

for ax in axes.flat:
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3, frameon=False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('spectral_density.pdf', bbox_inches='tight')
plt.show()
```
