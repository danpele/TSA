```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_spectral_density_gallery
# AR(1) and MA(1) theoretical spectra on linear and log scales
omega = np.linspace(0, np.pi, 500)
sigma2 = 1.0

def ar1_spectrum(phi, omega): return sigma2 / (1 - 2*phi*np.cos(omega) + phi**2)
def ma1_spectrum(theta, omega): return sigma2 * (1 + 2*theta*np.cos(omega) + theta**2)

fig, axes = plt.subplots(2, 2, figsize=(11, 7))

for phi, col in [(0.3, COLORS['blue']), (0.6, COLORS['orange']), (0.9, COLORS['red'])]:
    axes[0,0].plot(omega/np.pi, ar1_spectrum(phi, omega), color=col, lw=1.8, label=f'φ={phi}')
axes[0,0].set_title('AR(1) Spectral Density')
axes[0,0].set_xlabel('Frequency (×π)'); axes[0,0].set_ylabel('f(ω)')

for theta, col, ls in [(0.5, COLORS['green'], '-'), (-0.5, COLORS['purple'], '--')]:
    axes[0,1].plot(omega/np.pi, ma1_spectrum(theta, omega), color=col, lw=1.8, linestyle=ls, label=f'θ={theta}')
axes[0,1].set_title('MA(1) Spectral Density')
axes[0,1].set_xlabel('Frequency (×π)'); axes[0,1].set_ylabel('f(ω)')

for phi, col in [(0.3, COLORS['blue']), (0.6, COLORS['orange']), (0.9, COLORS['red'])]:
    axes[1,0].semilogy(omega/np.pi, ar1_spectrum(phi, omega), color=col, lw=1.8, label=f'φ={phi}')
axes[1,0].set_title('AR(1) — Log Scale')
axes[1,0].set_xlabel('Frequency (×π)'); axes[1,0].set_ylabel('log f(ω)')

for theta, col, ls in [(0.5, COLORS['green'], '-'), (-0.5, COLORS['purple'], '--')]:
    axes[1,1].semilogy(omega/np.pi, ma1_spectrum(theta, omega), color=col, lw=1.8, linestyle=ls, label=f'θ={theta}')
axes[1,1].set_title('MA(1) — Log Scale')
axes[1,1].set_xlabel('Frequency (×π)'); axes[1,1].set_ylabel('log f(ω)')

for ax in axes.flat:
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3, frameon=False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('ch12_spectral_density_gallery.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_spectral_density_gallery.png', bbox_inches='tight', dpi=150)
plt.show()
```
