```python
import os
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.filters.hp_filter import hpfilter
from statsmodels.tsa.filters.bk_filter import bkfilter
from statsmodels.tsa.filters.cf_filter import cffilter
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_gdp_bandpass
# Band-pass filtering of US real GDP
data = sm.datasets.macrodata.load_pandas().data
gdp = np.log(data['realgdp'].values)
year_q = data['year'].values + (data['quarter'].values - 1) / 4

cycle_hp, _ = hpfilter(gdp, lamb=1600)
cycle_bk = bkfilter(gdp, low=6, high=32, K=12)
cycle_cf, _ = cffilter(gdp, low=6, high=32)

T = len(gdp)
n_bk = len(cycle_bk)
start_bk = (T - n_bk) // 2

fig, ax = plt.subplots(figsize=(11, 4))
ax.plot(year_q, cycle_hp, color=COLORS['blue'],  lw=1.3, label='HP cycle (λ=1600)')
ax.plot(year_q[start_bk:start_bk+n_bk], cycle_bk, color=COLORS['red'], lw=1.3, label='BK cycle (6–32q, K=12)')
ax.plot(year_q, cycle_cf, color=COLORS['green'], lw=1.3, label='CF cycle (6–32q)')
ax.axhline(0, color='black', lw=0.8)
ax.set_title('Business Cycle Components of US Real GDP (BK & CF Band-Pass Filters)')
ax.set_xlabel('Year')
ax.set_ylabel('Cycle component (log GDP)')
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, frameon=False)
plt.tight_layout()
plt.savefig('ch12_gdp_bandpass.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_gdp_bandpass.png', bbox_inches='tight', dpi=150)
plt.show()
```
