**Description:** Band-pass filtering of GDP — Baxter-King and Christiano-Fitzgerald business cycle extraction

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.filters.hp_filter import hpfilter
from statsmodels.tsa.filters.bk_filter import bkfilter
from statsmodels.tsa.filters.cf_filter import cffilter

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
data = sm.datasets.macrodata.load_pandas().data
gdp = np.log(data['realgdp'].values)

# HP filter
cycle_hp, trend_hp = hpfilter(gdp, lamb=1600)

# BK filter (6-32 quarters, K=12)
cycle_bk = bkfilter(gdp, low=6, high=32, K=12)

# CF filter
cycle_cf, _ = cffilter(gdp, low=6, high=32)

fig, ax = plt.subplots(figsize=(11, 4))
T = len(gdp)
qtrs = np.arange(T)

# Align lengths
n_bk = len(cycle_bk)
start_bk = (T - n_bk) // 2

ax.plot(qtrs, cycle_hp, color=COLORS['blue'],   lw=1.3, label='HP cycle (λ=1600)')
ax.plot(qtrs[start_bk:start_bk+n_bk], cycle_bk, color=COLORS['red'],   lw=1.3, label='BK cycle (6–32q, K=12)')
ax.plot(qtrs, cycle_cf, color=COLORS['green'],  lw=1.3, label='CF cycle (6–32q)')
ax.axhline(0, color='black', lw=0.8)
ax.set_title('Business Cycle Components of US Real GDP')
ax.set_xlabel('Quarter (1959Q1 onwards)')
ax.set_ylabel('Cycle component (log)')
fig.patch.set_alpha(0); ax.set_facecolor('none')
ax.spines[['top','right']].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, frameon=False)
plt.tight_layout()
plt.savefig('gdp_bandpass.pdf', bbox_inches='tight')
plt.show()
```
