```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_fourier_decomposition
# DFT amplitude spectrum of a composite sinusoidal signal
N = 256
t = np.arange(N)
f1, f2, f3 = 5/N, 13/N, 30/N
x = (np.sin(2*np.pi*f1*t) +
     0.6*np.sin(2*np.pi*f2*t) +
     0.3*np.sin(2*np.pi*f3*t))

X = np.fft.rfft(x)
freqs = np.fft.rfftfreq(N)
amplitudes = np.abs(X) / N

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(t, x, color=COLORS['blue'], lw=1)
ax1.set_xlabel('Time (samples)')
ax1.set_ylabel('Amplitude')
ax1.set_title('Composite Signal (3 sinusoids)')

ax2.stem(freqs, amplitudes, linefmt=COLORS['blue'], markerfmt='o', basefmt='k-')
ax2.set_xlabel('Frequency (cycles/sample)')
ax2.set_ylabel('|X(f)| / N')
ax2.set_title('Amplitude Spectrum (FFT)')
for f_mark, label in zip([f1, f2, f3], ['f₁', 'f₂', 'f₃']):
    ax2.axvline(f_mark, color=COLORS['red'], lw=1, linestyle='--', alpha=0.6)

for ax in (ax1, ax2):
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('ch12_fourier_decomposition.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_fourier_decomposition.png', bbox_inches='tight', dpi=150)
plt.show()
```
