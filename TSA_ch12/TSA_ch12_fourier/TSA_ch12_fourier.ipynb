**Description:** Fourier decomposition of periodic signals

```python
import numpy as np
import matplotlib.pyplot as plt

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
N = 256
t = np.arange(N)
# Sum of three sinusoids
f1, f2, f3 = 5/N, 13/N, 30/N
x = (np.sin(2*np.pi*f1*t) +
     0.6*np.sin(2*np.pi*f2*t) +
     0.3*np.sin(2*np.pi*f3*t))

# FFT
X = np.fft.rfft(x)
freqs = np.fft.rfftfreq(N)
amplitudes = np.abs(X) / N

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(t, x, color=COLORS['blue'], lw=1)
ax1.set_xlabel('Time (samples)')
ax1.set_ylabel('Amplitude')
ax1.set_title('Composite Signal')

ax2.stem(freqs, amplitudes, linefmt=COLORS['blue'], markerfmt='o', basefmt='k-')
ax2.set_xlabel('Frequency (cycles/sample)')
ax2.set_ylabel('|X(f)| / N')
ax2.set_title('Amplitude Spectrum (FFT)')
for f_mark in [f1, f2, f3]:
    ax2.axvline(f_mark, color=COLORS['red'], lw=1, linestyle='--', alpha=0.6)

for ax in (ax1, ax2):
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('fourier_decomposition.pdf', bbox_inches='tight')
plt.show()
```
