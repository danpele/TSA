```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_periodogram_inconsistency
# Periodogram variance stays constant as N increases
rng = np.random.default_rng(42)
phi = 0.7
true_psd = lambda f: 1.0 / (1 - 2*phi*np.cos(2*np.pi*f) + phi**2)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, N, col in zip(axes, [128, 512, 2048], [COLORS['blue'], COLORS['orange'], COLORS['red']]):
    eps = rng.normal(0, 1, N + 50)
    x = np.zeros(N + 50)
    for t in range(1, N + 50):
        x[t] = phi * x[t-1] + eps[t]
    x = x[50:]
    freq, psd = periodogram(x)
    f_grid = np.linspace(0.01, 0.49, 300)
    ax.plot(freq[1:], psd[1:], color=col, lw=0.8, alpha=0.7, label=f'Periodogram N={N}')
    ax.plot(f_grid, true_psd(f_grid), color='black', lw=1.5, linestyle='--', label='True PSD')
    ax.set_title(f'N = {N}')
    ax.set_xlabel('Frequency'); ax.set_ylabel('PSD')
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=1, frameon=False)

fig.patch.set_alpha(0)
plt.suptitle('Periodogram Inconsistency: Variance Does Not Decrease with N', y=1.02)
plt.tight_layout()
plt.savefig('ch12_periodogram_inconsistency.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_periodogram_inconsistency.png', bbox_inches='tight', dpi=150)
plt.show()
```
