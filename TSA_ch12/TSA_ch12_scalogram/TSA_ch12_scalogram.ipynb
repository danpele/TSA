```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_scalogram_synthetic
# CWT scalogram of a chirp signal showing increasing frequency
N = 512
t = np.linspace(0, 10, N)
f_inst = 0.5 + 2.5 * t / 10
x = np.sin(2 * np.pi * f_inst * t)

fig, axes = plt.subplots(2, 1, figsize=(11, 7), gridspec_kw={'height_ratios': [1, 2.5]})
axes[0].plot(t, x, color=COLORS['blue'], lw=1)
axes[0].set_xlabel('Time (s)'); axes[0].set_ylabel('Amplitude')
axes[0].set_title('Chirp Signal (0.5 → 3 Hz)')
axes[0].spines[['top','right']].set_visible(False)
axes[0].set_facecolor('none')

wavelet = 'cmor1.5-1.0'
scales = np.geomspace(2, 128, 80)
coeffs, freqs_cwt = pywt.cwt(x, scales, wavelet, sampling_period=t[1]-t[0])
power = np.abs(coeffs)**2

im = axes[1].contourf(t, freqs_cwt, power, levels=30, cmap='YlOrRd')
plt.colorbar(im, ax=axes[1], label='Power')
axes[1].set_ylabel('Frequency (Hz)'); axes[1].set_xlabel('Time (s)')
axes[1].set_title('Scalogram (CWT power) — Chirp Signal')
axes[1].set_yscale('log')
axes[1].spines[['top','right']].set_visible(False)

fig.patch.set_alpha(0)
for ax in axes: ax.set_facecolor('none')
plt.tight_layout()
plt.savefig('ch12_scalogram_synthetic.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_scalogram_synthetic.png', bbox_inches='tight', dpi=150)
plt.show()
```
