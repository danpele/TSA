```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal.windows import dpss
```

```python
# Style configuration
COLORS = {
    'blue':   '#1A3A6E',
    'red':    '#DC3545',
    'green':  '#2E7D32',
    'orange': '#E67E22',
    'gray':   '#666666',
    'purple': '#8E44AD',
}

plt.rcParams.update({
    'axes.facecolor':     'none',
    'figure.facecolor':   'none',
    'savefig.transparent': True,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.grid':          False,
    'font.size':          9,
    'axes.labelsize':     9,
    'axes.titlesize':     10,
    'legend.fontsize':    8,
    'xtick.labelsize':    8,
    'ytick.labelsize':    8,
})
```

```python
# Chart: ch12_multitaper_dpss
# DPSS Slepian tapers and spectral concentration ratios
N = 256
NW = 4
K = 7
tapers, ratios = dpss(N, NW=NW, Kmax=K, return_ratios=True)

cmap_colors = [COLORS['blue'], COLORS['red'], COLORS['green'],
               COLORS['orange'], COLORS['orange'], COLORS['purple'], COLORS['gray']]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for i in range(K):
    axes[0].plot(tapers[i], color=cmap_colors[i], lw=1.3, label=f'k={i}, conc={ratios[i]:.4f}')
axes[0].set_title(f'DPSS Tapers (NW={NW}, K={K})')
axes[0].set_xlabel('Sample'); axes[0].set_ylabel('Taper amplitude')
axes[0].spines[['top','right']].set_visible(False)
axes[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4, frameon=False, fontsize=8)

axes[1].bar(np.arange(K), ratios, color=cmap_colors, edgecolor='none')
axes[1].set_title('Spectral Concentration Ratios')
axes[1].set_xlabel('Taper index k'); axes[1].set_ylabel('Concentration'); axes[1].set_ylim(0, 1.05)
for i, r in enumerate(ratios):
    axes[1].text(i, r + 0.01, f'{r:.4f}', ha='center', fontsize=7)
axes[1].spines[['top','right']].set_visible(False)

for ax in axes:
    ax.set_facecolor('none')
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('ch12_multitaper_dpss.pdf', bbox_inches='tight', dpi=150)
plt.savefig('ch12_multitaper_dpss.png', bbox_inches='tight', dpi=150)
plt.show()
```
