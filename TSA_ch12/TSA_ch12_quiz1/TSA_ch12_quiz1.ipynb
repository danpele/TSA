**Description:** Quiz 1 — spectral density identification

**Question:** Match each panel to: White Noise, AR(1) φ=0.8, MA(1) θ=-0.8, AR(2) with spectral peak.

```python
import numpy as np
import matplotlib.pyplot as plt

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
omega = np.linspace(0, np.pi, 500)
sigma2 = 1.0

spectra = {
    'A': np.ones(500),                                  # White noise
    'B': sigma2 / (1 - 2*0.8*np.cos(omega) + 0.8**2),  # AR(1) phi=0.8
    'C': sigma2 * (1 - 2*0.8*np.cos(omega) + 0.8**2),  # MA(1) theta=-0.8
    'D': None,                                           # AR(2) with peak
}
# AR(2) with complex roots near frequency pi/4
phi1, phi2 = 1.2, -0.9
spectra['D'] = sigma2 / np.abs(1 - phi1*np.exp(-1j*omega) - phi2*np.exp(-2j*omega))**2

labels = list('ABCD')
cols = [COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['purple']]

fig, axes = plt.subplots(1, 4, figsize=(14, 3.5))
for ax, lbl, col in zip(axes, labels, cols):
    ax.plot(omega/np.pi, spectra[lbl].real, color=col, lw=2)
    ax.set_title(f'Panel {lbl}')
    ax.set_xlabel('Frequency (×π)')
    ax.set_ylabel('f(ω)')
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.suptitle('Quiz 1: Which spectral density corresponds to which process?', y=1.02)
plt.tight_layout()
plt.savefig('quiz1_spectral_density.pdf', bbox_inches='tight')
plt.show()
print('Answer: A=White Noise, B=AR(1)φ=0.8, C=MA(1)θ=-0.8, D=AR(2) with peak')
```
