**Description:** Cross-spectral coherence between economic time series

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import coherence

COLORS = {'blue':'#1A3A6E','red':'#DC3545','green':'#2E7D32','amber':'#B5853F','orange':'#E6802E','purple':'#8E44AD','gray':'#666666'}
```

```python
rng = np.random.default_rng(7)
N = 512
delay = 3

# AR(2) process X
X = np.zeros(N + 50)
for t in range(2, N + 50):
    X[t] = 1.3*X[t-1] - 0.7*X[t-2] + rng.normal()
X = X[50:]

# Y = X shifted by 3 + noise
Y = np.concatenate([np.zeros(delay), X[:-delay]]) + 0.5*rng.normal(size=N)

f_coh, Cxy = coherence(X, Y, nperseg=128)

# Cross-spectrum for phase
from scipy.signal import csd, welch as welch_fn
f_csd, Pxy = csd(X, Y, nperseg=128)
phase = np.angle(Pxy, deg=True)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(f_coh, Cxy, color=COLORS['blue'], lw=1.8)
axes[0].axhline(0.5, color=COLORS['red'], linestyle='--', lw=1, label='0.5 threshold')
axes[0].set_title(f'Squared Coherence (delay={delay} periods)')
axes[0].set_xlabel('Frequency')
axes[0].set_ylabel('|C_xy(f)|²')
axes[0].set_ylim(0, 1.05)
axes[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=1, frameon=False)

axes[1].plot(f_csd, phase, color=COLORS['orange'], lw=1.3)
axes[1].axhline(0, color='black', lw=0.8)
axes[1].set_title('Cross-Spectral Phase (degrees)')
axes[1].set_xlabel('Frequency')
axes[1].set_ylabel('Phase (°)')

for ax in axes:
    ax.set_facecolor('none')
    ax.spines[['top','right']].set_visible(False)
fig.patch.set_alpha(0)
plt.tight_layout()
plt.savefig('coherence_example.pdf', bbox_inches='tight')
plt.show()
```
